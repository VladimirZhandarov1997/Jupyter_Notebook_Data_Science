### Найдём столбцы, в которых нужно поменять тип данных

**Распишем, в кахих слолбцах будем производить замену:**
1. year_of_release — год выпуска. Имеет тип float64, но это только год, значит меняем не на datetime,  а на int.
2. critic_score — оценка критиков (максимум 100). Так как у нас стобальная оценка, то дробная часть не несёт смысловой нагрузки, поэтому меняем float64 на int.
3. user_score — оценка пользователей (максимум 10). Так как у нас десятибальная оценка, то дробная часть важна, поэтому меняем object на float64.

```python
data['year_of_release'] = data['year_of_release'].fillna(-1).astype('int64')
data['critic_score'] = data['critic_score'].fillna(-1).astype('int64')
```

```python
def float_fun(row):
    try:
        return float(row)
    except:
        return 
```

```python
data['user_score'] = data['user_score'].apply(float_fun)
```

```python
data[['year_of_release','critic_score', 'user_score']].head()
```

```python
data.info()
```

```python
data.head(10)
```

Рассмотрим пустые значения в столбце с именем и жанром.

```python
data.isna().sum()
```

```python
data[data['name'].isna() | data['genre'].isna()]
```

Эти две строки содержат много нулевых значений, поэтому можем смело удалить.

Теперь обратим внимание на рейтенги.

Просто так удалить строчки с пропущенными значениями не можем (это практически треть всех данных).

Остались ещё нулевые значения. В столбце `rating` заменим NaN на undefined, а в столбце `user_score` заменим на `-1` аналогично столбцу `critic_score`.

```python
data['user_score'] = data['user_score'].fillna(-1)
```

```python
data['rating'] = data['rating'].fillna('undefined')
```

```python
data.info()
```

На этом закончили с обработкой пропусков. Назовем несколько причин возникновения пустых значений:
1. При переносе данных могли возникнуть ошибки.
2. Самое большое количество в столбцах с ретингами, из пункта 2.1 видно, что некотрые платформы имеют рейтинг всех игр, а некотрые ни одной. Возможно дело в самой платформе.
3. Также очень старые игры ил очень непопулярные игры могли быть не оценены.
4. Доля удалённых данных составила `1 - 16444/16715 = 0,0185` всего 1,85%.

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

* Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
* Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC разные.

```python
sample_XOne = data_platform_2016.query('platform == "XOne" and user_score != -1')['user_score']
sample_PC = data_platform_2016.query('platform == "PC" and user_score != -1')['user_score']

alpha = .05 

results_XOne_PC = st.ttest_ind(
    sample_XOne, 
    sample_PC, 
    equal_var = False)

print('p-значение:', results_XOne_PC.pvalue)

if results_XOne_PC.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Мы не опровергаем нулевую гипотезу, так как мы приняли уровень статической значимости в 5%, а вероятность равенста целых 54,9%.

### Средние пользовательские рейтинги жанров Action и Sports разные

* Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports одинаковые.
* Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.

```python
sample_Action = data_platform_2016.query('genre == "Action" and user_score != -1')['user_score']
sample_Sports = data_platform_2016.query('genre == "Sports" and user_score != -1')['user_score']

alpha = .05 

results_Action_Sports = st.ttest_ind(
    sample_Action, 
    sample_Sports, 
    equal_var = False)

print('p-значение:', results_Action_Sports.pvalue)

if results_Action_Sports.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Выходит и правда, чтобы рейтинги жанров Action и Sports совпали, должно сильно повезти. Но это практически не возможно, при вероятности в 10^(-20).

**Итого:**
* Мы формулировали гипотезы так, чтобы нулевая была сформулирована на равенство, тогда альтернативная была противоположна и было неравенство.
* Мы использовали двустореннюю гипотезу, так как нам по условию достаточно знать, равно или не равно, а вот больше или меньше нам не важно.
* Мы не смогли опровергнуть первую гипотезу и опровергли вторую.
